```python
#libraries
import pandas as pd
import numpy as np
import plotly.express as px
import sys
from bs4 import BeautifulSoup
import requests
from operator import pos

```

```python
#Function to Get and Set up Tables
def get_web_tables(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    tables = soup.find_all("table")
    df_list = pd.read_html(str(tables))
    df_list = df_list[0]
    df = pd.DataFrame (df_list, columns = ["Rk","Pos","Name","Age","G","PA","AB","R","H","2B","3B","HR","RBI","SB","CS","BB","SO","BA","OBP","SLG"])
    df = df[:-2]
    df = df.drop(df.index[9])
    df = df.fillna("-")
    df = df[df["Pos"].str.contains("P")==False]
    df = df[df["Name"].str.contains("Rank in 15 NL teams")==False]
    df = df[df["Name"].str.contains("Non-Pitcher Totals")==False]
    df = df[df["Name"].str.contains("Team Totals")==False]
    df["R"] = df["R"].astype(str).astype(int)
    df["RBI"] = df["RBI"].astype(str).astype(int)
    df["HR"] = df["HR"].astype(str).astype(int)
    Total_Runs = df["R"].astype(int)
    Total_Runs = Total_Runs.sum()
    df["TeamRuns"]= Total_Runs
    df["TeamValue"] = (df["R"]+df["RBI"]-df["HR"])/df["TeamRuns"]
    df = df.set_index("Name")

    return(df)
```

```python
#Function for finding the best player
def MVP(df):
  player = df["TeamValue"].idxmax()
  return(player)
```

```python
# joining teams
#National League East
NLEast = [ATL,MIA,NYM,PHI,WSN]
NLEast = pd.concat(NLEast)
 
#National League Central
NLCentral = [STL,MIL,CHC,PIT,CIN]
NLCentral = pd.concat(NLCentral)
 
#National League West
NLWest = [LAD,SD,SF,ARI,COL]
NLWest = pd.concat(NLWest)
 
#American League

#American League East
ALEast = [NYY,TOR,BAL,TB, BOS]
ALEast = pd.concat(ALEast)

#American League Central
ALCentral = [CLE, CHW, MIN, DET, KC]
ALCentral = pd.concat(ALCentral)

#American League West
ALWest = [HOU, SEA, TEX, LAA, OAK]
ALWest = pd.concat(ALWest)

#National League
NL = [NLEast,NLWest,NLCentral]
NL = pd.concat(NL)
NL["League"] = "NL"
#American League
AL = [ALEast, ALWest, ALCentral]
AL = pd.concat(AL)
AL["League"] = "AL"
#All of MLB
MLB = [NL,AL]
MLB = pd.concat(MLB)
```
